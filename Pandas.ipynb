```python
# --- 1. Create a DataFrame ---
data = {
    'City': ['NYC', 'London', 'Paris', 'Tokyo', 'Sydney'],
    'Population_M': [8.4, 8.9, 2.1, 13.9, 5.3],
    'Country': ['USA', 'UK', 'France', 'Japan', 'Australia'],
    'GDP_B': [1800, 700, 750, 1900, 350]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# --- 2. Data Selection and Filtering ---
# Select specific columns
cities_pop = df[['City']]
print("\nSelected Columns:")
print(cities_pop)
```

```python
# Filter rows where population is greater than 8 million
large_cities = df[df['Population_M'] > 8]
print("\nLarge Cities (Population > 8M):")
print(large_cities)
```

```python
# --- 3. Creating a New Column (Feature Engineering) ---
# Create a new column representing GDP per capita (approx)
df['GDP_per_Capita'] = (df['GDP_B'] / df['Population_M']).round(2)
print("\nDataFrame with New Column:")
print(df)
```

```python
# --- 4. Grouping and Aggregation ---
# Group by a categorical column (Country in a real dataset) and calculate the mean
# Using 'Country' as a proxy for demonstration purposes here
country_stats = df.groupby('Country')['Population_M'].mean()
print("\nMean Population by Country:")
print(country_stats)
```
