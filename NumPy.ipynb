```python
# --- 1. Create a NumPy Array ---
# Create a 3x3 array of random integers
data = np.random.randint(1, 10, size=(3, 3))
print("Original Array:")
print(data)
```

```python
# --- 2. Array Operations (Vectorization) ---
# Add a constant to every element
data_plus_10 = data + 10
print("\nArray + 10:")
print(data_plus_10)
```

```python
# Multiply two arrays element-wise
array_2 = np.array([[1, 0, 1], [0, 1, 0], [1, 0, 1]])
multiplied = data * array_2
print("\nElement-wise Multiplication:")
print(multiplied)
```

```python
# --- 3. Statistical Functions ---
print(f"\nMean of all elements: {data.mean():.2f}")
print(f"Sum of elements in each column (axis=0): {data.sum(axis=0)}")
```
