```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
# Set a style for better visuals
sns.set_theme(style="whitegrid")

# Create a sample DataFrame for plotting
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [150, 180, 220, 250, 300],
    'Expenses': [100, 110, 120, 130, 140]
}
sales_df = pd.DataFrame(data)
```

```python
# --- 1. Matplotlib: Basic Line Plot ---
plt.figure(figsize=(8, 4))
plt.plot(sales_df['Month'], sales_df['Sales'], marker='o', linestyle='-', color='b', label='Sales')
plt.plot(sales_df['Month'], sales_df['Expenses'], marker='x', linestyle='--', color='r', label='Expenses')

plt.title('Monthly Sales and Expenses (Matplotlib)')
plt.xlabel('Month')
plt.ylabel('Amount (in thousands)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# --- 2. Seaborn: Categorical Bar Plot ---
# Create a scatterplot of Sales vs. Expenses
plt.figure(figsize=(6, 6))
sns.scatterplot(x='Expenses', y='Sales', data=sales_df, s=150, color='purple')
plt.title('Sales vs. Expenses (Seaborn Scatter)')
plt.show()
```

```python
# --- 3. Seaborn: Distribution Plot (using the Cities DataFrame from above) ---
# Re-using the Cities DataFrame structure for demonstration
cities_df = pd.DataFrame({
    'City': ['NYC', 'London', 'Paris', 'Tokyo', 'Sydney', 'LA', 'Chicago'],
    'Population_M': [8.4, 8.9, 2.1, 13.9, 5.3, 3.9, 2.7]
})
```

```python
plt.figure(figsize=(8, 5))
# Use a bar plot to visualize a single distribution
sns.barplot(x='City', y='Population_M', data=cities_df, palette='viridis')
plt.title('City Populations (Seaborn Bar Plot)')
plt.ylabel('Population (Millions)')
plt.xlabel('City')
plt.xticks(rotation=45) # Rotate city names for readability
plt.tight_layout()
plt.show()
```
